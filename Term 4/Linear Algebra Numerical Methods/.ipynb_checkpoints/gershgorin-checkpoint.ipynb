## Gershgorin Circle Theorem

Consider a matrix $\boldsymbol{A} \in \mathbb{R}^{n\times n}$. Consider circles on the complex plane:
$$
\mathcal{K}_i = \left\{ z \in \mathbb{C}: |z-a_{i,i}| \leq \sum_{i \neq j}|a_{i,j}| \right\}, \; i = \overline{1,n}
$$

Then $\sigma(\boldsymbol{A}) \subset \bigcup_{i=1}^n \mathcal{K}_i$

**Task**. 

1. Take arbitrary $4 \times 4$ matrix $\boldsymbol{A}$. Build Gershgorin Circles and localize spectrum. Use Gershgorin Circle Theorem for $\boldsymbol{A}$ and $\boldsymbol{A}^{\top}$
2. Find spectrum of this matrix and verify that it indeed belongs to built circles. 

**Solution**. Consider a matrix
$$
\boldsymbol{A} = \begin{bmatrix}
1 & 4 & 3 & 4 \\
5 & 6 & 8 & -5 \\
4 & -6 & 4 & 12 \\
-1 & 3 & 9 & 5
\end{bmatrix}
$$

Then we have the following circles:
$$
\mathcal{K}_1 = \{ z \in \mathbb{C}: |z-1| \leq 11 \},\;\mathcal{K}_2 = \{ z \in \mathbb{C}: |z-6| \leq 18 \},
$$
$$
\mathcal{K}_3 = \{ z \in \mathbb{C}: |z-4| \leq 22 \}, \; \mathcal{K}_4 = \{ z \in \mathbb{C}: |z-5| \leq 13 \}
$$

Transposed matrix:
$$
\boldsymbol{A}^{\top} = \begin{bmatrix}
1 & 5 & 4 & -1 \\
4 & 6 & -6 & 3 \\
3 & 8 & 4 & 9 \\
4 & -5 & 12 & 5
\end{bmatrix}
$$

Applying the same method for $\boldsymbol{A}^{\top}$ yields:
$$
\mathcal{K}_1' = \{ z \in \mathbb{C}: |z-1| \leq 10 \}, \; \mathcal{K}_2' = \{ z \in \mathbb{C}: |z-6| \leq 13 \}
$$
$$
\mathcal{K}_3' = \{ z \in \mathbb{C}: |z-4| \leq 20 \}, \; \mathcal{K}_4' = \{ z \in \mathbb{C}: |z-5| \leq 21 \}
$$

```python
def print_gershgorin_circles(A, ax, color='blue'):
    """
    Function that draws gershgorin circles and eigenvalues of a given matrix A
    
    Input:
    A - matrix
    ax - plt.figure
    color - (optional) color of circles
    
    Output:
    Drawn figure with circles and eigenvalues
    """
    n = len(A) # Dimensionality of A
    
    radiuses = np.sum(np.abs(A),axis=1) - np.abs(np.diagonal(A))
    eigenvalues = np.linalg.eigvals(A)
    for i in range(n):
        circle = plt.Circle((A[i,i], 0), radiuses[i], color=color, fill=False)
        ax.add_patch(circle)
        ax.plot(eigenvalues[i].real, eigenvalues[i].imag, 'r+', markersize=12, mew=3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot()
ax.set_aspect('equal')
ax.grid()

A = np.array([
    [1,5,4,-1],
    [4,6,-6,3],
    [3,8,4,9],
    [4,-5,12,5]])

# Green circles are Gershgorin circles for a tranposed matrix, whereas blue ones for a non-tranposed one
print_gershgorin_circles(np.transpose(A), ax, color='green')
print_gershgorin_circles(A, ax)
```
